지도학습:
예측할 새로운 x 데이터가 주어졌을 때 모델 학습을 통해 찾아낸 관계식에 x를 대입하면 목표 변수인 y를 예측할 수 있다

일차함수 관계식 찾기

문제파악

```python
x = [-3, 31,-11,4,0,22,-2,-5,-25,-14]
y = [-2,32,-10,5,1,23,-1,-4,-24,-13]
```

데이터 탐색

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.show()
```

데이터 전처리

```python
import pandas as pd
df = pd.DataFrame({'X':x,'Y':y})
df.shape  #몇행 몇열
```

```python
df.head() #첫 5개 추출
```

```python
df.tail()
```

```python
train_features = ['X']
target_cols = ['Y']
X_train = df.loc[:,train_features]
y_train = df.loc[:,target_cols]

print(X_train)
print(y_train)
```

모델 학습

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression() #모델 인스턴스 객체 생성한 후 lr변수에 저장
lr.fit(X_train, y_train)  #입력 데이터를 모델에 전달하여 학습시키는 함수 명령
```

```python
lr.coef_,lr.intercept_  #coef_ : X변수의 회구계수(기울기), intercept_ 상수항(y절편)
```

```python
print(lr.coef_)
print(lr.intercept_)

print(lr.coef_[0][0]) #입력 데이터가 2차원 구조이므로 회귀계수도 2차원 구조로 반환
print(lr.intercept_[0])
#y=x+1
```

예측

```python
import numpy as np
#학습에 사용한 X_train이 2차원 구조이므로, predict 메소드의 입력값은 2차원 구조를 가져야함
#numpy의 array함수로 숫자 11을 배열로 변환한 후 reshape 메소드를 적용하여 1행 1열 형태의 2차원 구조로 변형
X_new = np.array(11).reshape(1,1)
print(X_new)
lr.predict(X_new)
```

```python
#11부터 15까지 1의 간격을 두고 1차원 배열을 만든 후
#reshape메소드를 통해 n행 1열의 형태의 2차원 구조로 변환!
#-1은 크기가 정해지지 않았다는 뜻
X_test = np.arange(11,16,1).reshape(-1,1)

#11부터 15까지 1의 간격으로 5개 행 필요하니 -1  ->  5로 알아서 해석함
X_test
```

```python
y_pred = lr.predict(X_test)
y_pred
```
